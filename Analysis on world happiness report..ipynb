```python
wh1 = wh.rename(columns={'Happiness Rank':'happiness','Happiness Score':'score','Economy (GDP per Capita)':'economy',
                         'Health (Life Expectancy)':'health','Trust (Government Corruption)':'trust','Dystopia Residual':'dystopia'})
```

# Plot b/w Happiness Score and GDP per Capita

```python
plot.scatter(x=wh1.happiness,y=wh1.economy)
plot.title('Happiness Score vs GDP')
plot.xlabel('Happiness Score')
plot.ylabel('GDP')
```

```python
wh1.groupby('score')[['economy']].mean().head(10)
```

```python
plot.rcParams["figure.figsize"] = (14,9)
plot.title('Happiness Score vs GDP', size=20)
##plot.scatter(x=wh1.happiness,y=wh1.economy)
sns.scatterplot(x = wh1.happiness, y = wh1.economy, hue = wh1.Region, s=150)
plot.legend(fontsize = '10')
plot.xlabel("Happiness Score", size=15)
plot.ylabel("GDP", size=15)
sns.set_style('darkgrid')
```

# GDP by Region

```python
ax=wh1.groupby('Region')['economy'].sum()
ax
```

```python
ax.plot.pie(autopct='%0.1f%%')
plot.ylabel(' ')
```

# Total Countries in all regions

```python
ax = wh1.groupby('Region')[['Country']].count()
ax
```

```python
ax.plot.pie(autopct='%0.1f%%')
plot.ylabel(' ')
```

# Perception of Freedom in various Regions

```python
ax = wh1.groupby('Region')[['Freedom']].mean()
ax.plot.barh(color = 'g')
plot.title('Perception of Freedom in various Regions',size=20)
plot.xlabel('Freedom Index', size=15)
plot.ylabel('Regions', size=15)
sns.set_style('darkgrid')
plot.grid(True, color='g') 
```

# Perception of Corruptiopn in various Regions

```python
ax = wh1.groupby('Region')[['trust']].mean()
ax.plot.barh(color='y')
plot.title("Region Vs Corruption", size=20)
plot.ylabel('Region',size=15)
plot.xlabel('Corruption',size=15)
sns.set_style('darkgrid')
plot.grid(True, color='k') 
```

# Top 10 and Bottom 10 happiest countries Life Expectancy

```python
ax = wh1.sort_values(by = 'score', ascending= False).drop_duplicates('Country')
ax
```

```python
top_10 = ax.head(10)
top_10
```

```python
bottom_10 = ax.tail(10)
bottom_10
```

# Top 10 Life Expectancy

```python

sns.barplot(x=top_10.Country, y= top_10.health)
plot.xlabel("Country Name", size=20)
plot.ylabel("Life Expectancy", size=20)
plot.grid(True)
sns.set_style("darkgrid")
```

# Bottom 10 Life Expectancy

```python
sns.barplot(x=bottom_10.Country, y= bottom_10.health)
plot.xlabel("Country Name",size=20)
plot.ylabel("Life Expectancy", size=20)
plot.grid(True)
sns.set_style("darkgrid")
plot.xticks(rotation = 20, ha="right")
```

# Freedom To Make Life Choices

```python
plot.rcParams["figure.figsize"] = (17,8)
sns.scatterplot(x=wh1.Freedom,y=wh1.score,hue=wh1.Region, s=150)
plot.xlabel('Freedom',size=15)
plot.ylabel('Happiness Score',size=15)
plot.title("Happiness Score Vs Freedom", size=20)
```

# Countries with most Perception of Corruption

```python
a = wh1.sort_values(by='trust').drop_duplicates('Country').tail(10)
sns.barplot(x= a.Country, y=a.trust)
plot.title("Countries with most Perception of Corruption")
plot.xlabel("Countries", size = 13)
plot.ylabel("Corruption", size = 13)
```

# Corruption vs Happiness

```python
plot.rcParams["figure.figsize"]=(20,8)
sns.scatterplot(x=wh1.score , y=wh1.trust, hue=wh1.Region, s= 150)
plot.xlabel("Happiness Score", size=14)
plot.ylabel("Corruption", size=14 )
```

# Is Happiness Affected by Region?

```python
ax = wh1.groupby('Region')[['happiness']].mean()
ax.plot.barh(color = 'g')
plot.title('Happiness affected by Regions',size=20)
plot.ylabel('Region', size=15)
plot.xlabel('Happiness', size=15)
sns.set_style('darkgrid')
plot.grid(True, color='g') 

#By the observation we found happiness is affected by Region.
```

# Did the happiness score change significantly over the past three years?

```python
ax = wh1.groupby('Year')[['score']].mean()
ax
```

```python
ax.plot.bar(color='c')
plot.title('Observing Happiness Score by Year',size=20)
plot.xlabel('Year', size=15)
plot.ylabel('Happiness Score', size=15)
sns.set_style('darkgrid')
plot.grid(True, color='g') 

# By above observation we obserb there is no major changes in happiness score.
```
